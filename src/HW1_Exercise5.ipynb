```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Exercise 5-1

```python
def bm_path_generation(step_size,T, N, mu = 0, sigma = 1, B0 = 0):
    # T: maturity
    # N: number of simulation
    dt = int(T/step_size)
    bm_paths = np.zeros((N,dt))
    for i in range(N):
        for j in range(1,dt):
            bm_paths[i][j] = bm_paths[i][j-1] + np.sqrt(step_size)*np.random.normal(mu,sigma)
    
    return bm_paths

ex1 = bm_path_generation(1/500, 3, 1)
plt.plot(ex1[0])
```
