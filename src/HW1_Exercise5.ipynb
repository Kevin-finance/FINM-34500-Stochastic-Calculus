```python
def bm_path_generation(step_size,T, N, mu = 0, sigma =1, B0 = 0):
    # T: maturity
    # N: number of simulation
    dt = int(T/step_size)
    bm_paths = np.zeros((N,dt))
    for i in range(N):
        for j in range(1,dt):
            bm_paths[i][j] = bm_paths[i][j-1] + np.sqrt(1/step_size)*np.random.normal(mu,sigma)
    
    return bm_paths

temp = bm_path_generation(1/500,3, 1)
print(temp)
    
    
    
    
    # dt = T/N
    # t = np.linspace(0, T, N+1)
    # W = np.zeros(N+1)
    # S = np.zeros(N+1)
    # S[0] = S0
    # for i in range(1, N+1):
    #     W[i] = W[i-1] + np.sqrt(dt)*np.random.normal(0, 1)
    #     S[i] = S0*np.exp((mu - 0.5*sigma**2)*t[i] + sigma*W[i])
    # return S
```
